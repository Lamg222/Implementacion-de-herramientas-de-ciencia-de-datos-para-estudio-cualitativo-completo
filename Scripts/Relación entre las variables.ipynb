Con esto sólo obtenemos aquellos registros con la respuésta "Sí" en todas la experiencias. Lo que buscamos es que estén los registros que han indicado que sí para cada columna, por lo tanto, se tendrá que almacenar en DF distintos, para cada una de las variables.

```python
#Para la variable inicial
df_filtrado_1 = df[df.iloc[:, [59]].apply(lambda x: (x == 'Sí').all(), axis=1)]
df_filtrado_1.head()
```

```python
#Lo hacemos de manera iterativa para cada una de las variable
columnas = df.columns[59:84]  # Seleccionamos las columnas de interés

# Creamos un diccionario para almacenar los DataFrames filtrados
dataframes_filtrados = {}

# Iteramos sobre las columnas
for i, col in enumerate(columnas, start=1):
    # Creamos el DataFrame filtrado para cada columna
    df_filtrado = df[df[col] == 'Sí']
    # Guardamos el DataFrame filtrado en el diccionario
    dataframes_filtrados[f'df_filtrado_{i}'] = df_filtrado

# Ahora dataframes_filtrados contiene un DataFrame filtrado para cada columna individual
```

```python
# Accedemos a cada DataFrame y los convertimos en DataFrames individuales
for key, value in dataframes_filtrados.items():
    globals()[key] = pd.DataFrame(value)

# Ahora tienes DataFrames individuales df_filtrado_1, df_filtrado_2, etc.
```

### Tabla de contingencia

```python
# DF y las variables de experiencia (59 a la 83)
columnas = df.columns[59:84]  # Seleccionamos las columnas de interés

# Creamos una lista para almacenar los DataFrames de cada columna
dataframes = []

# Iteramos sobre las columnas
for col in columnas:
    # Creamos un DataFrame temporal para cada columna con los recuentos de "Sí" y "No"
    df_temp = pd.DataFrame(df[col].value_counts()).T
    # Asignamos el nombre de la columna como índice del DataFrame temporal
    df_temp.index = [col]
    # Añadimos los DataFrames temporales a la lista
    dataframes.append(df_temp)

# Concatenamos los DataFrames de cada columna en un único DataFrame
tabla_contingencia = pd.concat(dataframes, axis=0)

# Reemplazamos los valores NaN por 0
tabla_contingencia = tabla_contingencia.fillna(0)

# Mostramos la tabla de contingencia
print(tabla_contingencia)
```

```python
import re
columnas = df.columns[59:84]  # Seleccionamos las columnas de interés

# Creamos una lista para almacenar los DataFrames de cada columna
dataframes = []

# Iteramos sobre las columnas
for col in columnas:
    # Utilizamos expresiones regulares para extraer el texto entre corchetes
    nombre_fila = re.search('\[(.*?)\]', col).group(1)
    # Creamos un DataFrame temporal para cada columna con los recuentos de "Sí" y "No"
    df_temp = pd.DataFrame(df[col].value_counts()).T
    # Asignamos el nombre extraído como índice del DataFrame temporal
    df_temp.index = [nombre_fila]
    # Añadimos los DataFrames temporales a la lista
    dataframes.append(df_temp)

# Concatenamos los DataFrames de cada columna en un único DataFrame
tabla_contingencia = pd.concat(dataframes, axis=0)

# Reemplazamos los valores NaN por 0
tabla_contingencia = tabla_contingencia.fillna(0)

# Mostramos la tabla de contingencia
print(tabla_contingencia)
```

```python
columnas = df.columns[59:84]  # Seleccionamos las columnas de interés

# Creamos una lista para almacenar los DataFrames de cada combinación de respuestas
dataframes = []

# Iteramos sobre las columnas
for col in columnas:
    # Creamos una tabla de contingencia entre la variable 29 y la columna actual
    tabla_contingencia = pd.crosstab(index=df['¿En qué sustancia(s) psicodélica(s) se basarán tus respuestas? '], columns=df[col], normalize='columns')
    # Añadimos la tabla de contingencia a la lista
    dataframes.append(tabla_contingencia)

# Concatenamos los DataFrames de cada combinación de respuestas en un único DataFrame
tabla_completa = pd.concat(dataframes, axis=1)

# Reemplazamos los valores NaN por 0
tabla_completa = tabla_completa.fillna(0)

# Mostramos la tabla completa
print(tabla_completa)
```
