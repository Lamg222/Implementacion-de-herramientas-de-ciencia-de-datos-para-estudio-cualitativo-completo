```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Generar datos de ejemplo
np.random.seed(0)
X = np.linspace(-10, 10, 1000).reshape(-1, 1)
y = (X[:, 0] > 0).astype(int)
```

```python
# Crear y ajustar el modelo de regresión logística
model = LogisticRegression()
model.fit(X, y)
```

```python
# Calcular las probabilidades predichas
y_pred_prob = model.predict_proba(X)[:, 1]
```

```python
# Graficar los datos y la curva sigmoide
plt.figure(figsize=(10, 6))
plt.plot(X, y_pred_prob, label='Curva Sigmoide', color='red')
plt.scatter(X, y, label='Datos', color='blue', s=10)
plt.xlabel('Variable Independiente')
plt.ylabel('Probabilidad')
plt.title('Regresión Logística - Curva Sigmoide')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Generar un rango de valores para X
X = np.linspace(-5, 5, 100).reshape(-1, 1)

# Crear datos binarios para ajustar el modelo
y = (X[:, 0] > 0).astype(int)

# Crear y ajustar el modelo de regresión logística
model = LogisticRegression()
model.fit(X, y)

# Calcular las probabilidades predichas
y_pred_prob = model.predict_proba(X)[:, 1]
```

```python
# Graficar la curva sigmoide
fig, ax = plt.subplots(figsize=(10, 8))
plt.plot(X, y_pred_prob, label='Curva Sigmoide', color='red')

# Dibujar una línea horizontal a partir del valor 0.5
ax.axhline(y=0.5, color='g', linestyle='--')

plt.xlabel('x', fontweight= "bold")

# Dibujar una línea horizontal a partir del valor 0.5
ax.axhline(y=0.5, color='g', linestyle='--')

ax.set_yticks(np.arange(0, 1.1, 0.1))


plt.ylabel('Probabilidad', fontweight= "bold")
plt.title('Regresión Logística - Curva Sigmoide')
plt.legend()

# Agregar anotaciones al lado del gráfico
ax.text(6, .55, 'P(y) > 0.5', va='center', fontweight='bold')
ax.text(6, .5, 'P(y) = 0.5', va='center', fontweight='bold')
ax.text(6, .45, 'P(y) < 0.5', va='center', fontweight='bold')

plt.show()
```
