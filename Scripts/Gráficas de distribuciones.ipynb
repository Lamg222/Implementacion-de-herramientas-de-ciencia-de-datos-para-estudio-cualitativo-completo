**Chi-cuadrada**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Número de grados de libertad
df = [1, 2, 4, 5, 8, 10, 15]

# Rango de valores para el eje x
x = np.linspace(0, 30, 100)

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar las gráficas Chi-cuadradas
for d in df:
    y = chi2.pdf(x, d)
    ax.plot(x, y, label=f"df={d}")

# Personalizar el gráfico
ax.set_title("Gráficas Chi-cuadradas")
ax.set_xlabel("Valor")
ax.set_ylabel("Densidad de probabilidad")
ax.legend()
ax.grid(True)

# Mostrar el gráfico
plt.show()
```

**Hipergeométricas (para distribución de Fisher)**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

def plot_hypergeometric_distribution(N, n, N1, num_samples=100):
    """
    Grafica la distribución hipergeométrica.
    
    Parámetros:
    N (int): Tamaño de la población.
    n (int): Número de elementos seleccionados de la población.
    N1 (int): Número de elementos de interés en la población.
    num_samples (int): Número de muestras a generar (predeterminado: 100).
    """
    # Crear un objeto de distribución hipergeométrica
    dist = hypergeom(N, N1, n)
    
    # Generar muestras aleatorias
    samples = dist.rvs(num_samples)
    
    # Calcular la probabilidad de cada valor posible
    x = np.arange(n + 1)
    pmf = dist.pmf(x)
    
    # Crear el gráfico
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(x, pmf, 'bo', ms=8, label='Distribución hipergeométrica')
    ax.vlines(x, 0, pmf, colors='b', lw=2, alpha=0.5)
    ax.legend(loc='best', frameon=False)
    
    # Configurar el título y etiquetas de los ejes
    ax.set_title(f'Distribución Hipergeométrica (N={N}, n={n}, N1={N1})')
    ax.set_xlabel('Número de elementos seleccionados en la muestra')
    ax.set_ylabel('Probabilidad')
    
    # Mostrar el gráfico
    plt.show()

# Ejemplo de uso
N = 100  # Tamaño de la población
n = 30   # Número de elementos seleccionados de la población
N1 = 20  # Número de elementos de interés en la población

plot_hypergeometric_distribution(N, n, N1)
```
