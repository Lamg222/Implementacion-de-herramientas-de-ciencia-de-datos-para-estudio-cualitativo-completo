**Chi-cuadrada**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Nmero de grados de libertad
df = [1, 2, 4, 5, 8, 10, 15]

# Rango de valores para el eje x
x = np.linspace(0, 30, 100)

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar las grficas Chi-cuadradas
for d in df:
    y = chi2.pdf(x, d)
    ax.plot(x, y, label=f"df={d}")

# Personalizar el grfico
ax.set_title("Grficas Chi-cuadradas")
ax.set_xlabel("Valor")
ax.set_ylabel("Densidad de probabilidad")
ax.legend()
ax.grid(True)

# Mostrar el grfico
plt.show()
```

**Hipergeomtricas (para distribucin de Fisher)**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

def plot_hypergeometric_distribution(N, n, N1, num_samples=100):
    """
    Grafica la distribucin hipergeomtrica.
    
    Parmetros:
    N (int): Tamao de la poblacin.
    n (int): Nmero de elementos seleccionados de la poblacin.
    N1 (int): Nmero de elementos de inters en la poblacin.
    num_samples (int): Nmero de muestras a generar (predeterminado: 100).
    """
    # Crear un objeto de distribucin hipergeomtrica
    dist = hypergeom(N, N1, n)
    
    # Generar muestras aleatorias
    samples = dist.rvs(num_samples)
    
    # Calcular la probabilidad de cada valor posible
    x = np.arange(n + 1)
    pmf = dist.pmf(x)
    
    # Crear el grfico
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(x, pmf, 'bo', ms=8, label='Distribucin hipergeomtrica')
    ax.vlines(x, 0, pmf, colors='b', lw=2, alpha=0.5)
    ax.legend(loc='best', frameon=False)
    
    # Configurar el ttulo y etiquetas de los ejes
    ax.set_title(f'Distribucin Hipergeomtrica (N={N}, n={n}, N1={N1})')
    ax.set_xlabel('Nmero de elementos seleccionados en la muestra')
    ax.set_ylabel('Probabilidad')
    
    # Mostrar el grfico
    plt.show()

# Ejemplo de uso
N = 100  # Tamao de la poblacin
n = 30   # Nmero de elementos seleccionados de la poblacin
N1 = 20  # Nmero de elementos de inters en la poblacin

plot_hypergeometric_distribution(N, n, N1)
```
